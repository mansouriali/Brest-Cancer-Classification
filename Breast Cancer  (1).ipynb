### Important Note:
As we can see, the datset has 569 samples with 30 features; its label is binary and relatively balanced. 62.74% are positive and 37.26 are negative cases.
We assume that EDA has been done and we are ready to jump into the model training. We want to compare the performance of different classification algorithms including:

    1) Support Vector Machine
    2) Logistic Regression
    3) Decision Tree
    4) Random Forest
    5) Gradient Boosting Tree

The next steps are:

    1) Dividing the data into Training and Testing sets(Train 85%-Test 15% split)
    2) Scaling the data using StandardScaler
    3) Defining the model and fitting the data; (Cross Validation will be used to find the best parameters)
    4) Calculating the performance metrics


```python
#Step1:
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
#Step2:
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
#Step3:
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

```python
#Step3:
from sklearn.model_selection import GridSearchCV
```

SVC Model

```python
svc_model=SVC()
```

```python
svc_param={'kernel':['rbf','linear'],
      'C':[0.01,0.1,0.5,.9,.99,1],
      'gamma':['scale','auto']}
```

```python
svc_grid_model=GridSearchCV(svc_model,svc_param)
```

```python
svc_grid_model.fit(X_train,y_train)
```

```python
svc_best_model=svc_grid_model.best_estimator_
```

```python
svc_best_model.get_params()
```

Logistic Regression Model

```python
log_model=LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
log_model.get_params()
```

Decision Tree Model

```python
tree_model=DecisionTreeClassifier()
```

```python
tree_param={'criterion':['gini','entropy'],
            'max_depth':[3,4,5,6]}
```

```python
tree_grid_model=GridSearchCV(tree_model,tree_param)
```

```python
tree_grid_model.fit(X_train,y_train)
```

```python
tree_best_model=tree_grid_model.best_estimator_
```

```python
tree_best_model.get_params()
```

Random Forest Model

```python
rf_model=RandomForestClassifier()
```

```python
rf_param={'n_estimators':[64,100,128],
         'max_features':['auto','sqrt','log2'],
         'bootstrap':[True,False]}
```

```python
rf_grid_model=GridSearchCV(rf_model,rf_param)
```

```python
rf_grid_model.fit(X_train,y_train)
```

```python
rf_best_model=rf_grid_model.best_estimator_
```

```python
rf_best_model.get_params()
```

Gradient Boosting Tree Model

```python
gbt_model=GradientBoostingClassifier()
```

```python
gbt_param={'n_estimators':[64,100,128],
          'max_features':['auto','sqrt','log2'],
          'learning_rate':[0.1,0.5,1]}
```

```python
gbt_grid_model=GridSearchCV(gbt_model,gbt_param)
```

```python
gbt_grid_model.fit(X_train,y_train)
```

```python
gbt_best_model=gbt_grid_model.best_estimator_
```

```python
gbt_best_model.get_params()
```

```python
#Step4:
from sklearn.metrics import classification_report,plot_confusion_matrix
```

```python
#in order to make it easier to calculate the performance metrics, we create a function:
def performance_calculator(model,X_test,y_test):
    y_pred=model.predict(X_test)
    print(classification_report(y_test,y_pred))
    plot_confusion_matrix(model,X_test,y_test)
```

```python
performance_calculator(svc_best_model,X_test,y_test)
plt.savefig('SVC report.jpg')
```

```python
performance_calculator(log_model,X_test,y_test)
plt.savefig('logistic regression report.jpg')
```

```python
performance_calculator(tree_best_model,X_test,y_test)
plt.savefig('Decision tree report.jpg')
```

```python
performance_calculator(rf_best_model,X_test,y_test)
plt.savefig('random forest report.jpg')
```

```python
performance_calculator(gbt_best_model,X_test,y_test)
plt.savefig('gradient boosting tree report.jpg')
```

comparing the results, SVC and LogisticRegression achieve better performance.
In different tree models, RandomForest has performed better.
